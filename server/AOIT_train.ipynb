```python
import pandas as pd
from dateutil import parser
import numpy as np
data = pd.read_table('./whole_data.csv',sep=',')
```

```python
data = data[:-811]
# data = data.reset_index()
```

```python
value = []
temp = 0
count= 0
for i in range(len(data)):
    count+=1
    temp += data['Value'][i]        
    if(count==1):
        value.append(temp)
        count = 0
        temp=0
```

```python
value
```

```python
hour_data = []
temp = 0
count= 0
for i in range(len(value)):
    count+=1
    temp += data['Value'][i]
    if(count==60):
        hour_data.append(temp)
        count = 0
        temp=0
```

```python
hour_data
```

```python
hour = []
ct = 0
for i in range(len(hour_data)):
    hour.append(ct)
    ct += 1
    if(ct==24):
        ct = 0
```

```python
hour[0]
```

```python
day = []
ct = 0
for i in range(len(hour_data)):
    day.append(data['TimeStamp'][i*60][8:10])
```

```python
day
```

```python
x = np.empty([len(hour_data)-24, 26], dtype = float)
y = np.empty([len(hour_data)-24, 1], dtype = float)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from joblib import dump, load
for k in range(24,25):
    x = np.empty([len(hour_data)-24-k, 26])
    y = np.empty([len(hour_data)-24-k, 1])
    for i in range(x.shape[0]):
        x[i][0] = day[i]
        x[i][1] = hour[i]
        x[i][2:] = hour_data[i:i+24]
        y[i] = hour_data[i+k]
    lr = LinearRegression()
    lr.fit(x, y)
    lr.score(x,y)
    print(k,"= ",lr.score(x,y))
#dump(lr, 'lrmodel.joblib')
```
